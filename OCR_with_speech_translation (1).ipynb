```python
from googletrans import Translator
import easyocr
from gtts import gTTS #text to speech
from IPython.display import Audio
import PIL
from PIL import ImageDraw
```

```python
reader=easyocr.Reader(["mr","hi"]) #downloading hindi and marathi module
translator=Translator()
```

```python
image=PIL.Image.open("nana.png")
image
```

```python
text=reader.readtext("nana.png",detail=0) #extracting texts from image
```

```python
text_comb=" ".join(text) #joining all the sentences[lists] into one sentence[list]
text_comb
```

```python
print(translator.detect(text_comb)) #detecting the language from the image
```

## Translating from Marathi to Hindi

### Text

```python
text_hi =translator.translate(text_comb,src="mr",dest="hi")
print(text_hi.text)
```

### Audio

```python
hindi_text=gTTS(text_hi.text,lang="hi") #language here is the accent
hindi_text.save('hindi_audio.mp3')
```

```python
Audio('hindi_audio.mp3',autoplay=True)
```

## Translating from Marathi to English

```python
text_en = translator.translate(text_comb,src="mr") #as english is the default module for translation I did not give destination here 
print(text_en.text)
```

```python
eng_text=gTTS(text_en.text) #I have used the default us english accent
eng_text.save("english_audio.mp3")
```

```python
Audio("english_audio.mp3",autoplay=True)
```

## In french accent

It is the same english text but in a french accent the translation is the same the text has not been changed only the audio is different

```python
eng_text=gTTS(text_en.text,lang="fr") 
eng_text.save("english_text_french_accent_audio.mp3")
```

```python
Audio("english_text_french_accent_audio.mp3",autoplay=True)
```
